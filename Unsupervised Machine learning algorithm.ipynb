```python
#importing the library
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets

#Load the "Iris Dataset" 
iris=datasets.load_iris()
iris_df=pd.DataFrame(iris.data,columns=iris.feature_names)
iris_df.head()
```

```python
x = iris_df.iloc[:, [0, 1, 2, 3]].values

from sklearn.cluster import KMeans
wcss = []             #within cluster sum of squares 

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++',             
                    max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
    
# Plotting the results onto a line graph, 
# `allowing us to observe 'The elbow'
plt.plot(range(1, 11), wcss,'-o')
plt.title('The elbow method for finding the optimum number of clusters')
plt.xlabel('Number of clusters "K"')
plt.ylabel('within Cluster sum of square ') 
plt.show()
```

```python
# Applying kmeans to the dataset / Creating the kmeans classifier
kmeans = KMeans(n_clusters = 3, init = 'k-means++',
                max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(x)
```

```python
# Visualising the clusters 
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1],          #for sepal length and sepal width
            s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], 
            s = 100, c = 'yellow', label = 'Iris-versicolour')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1],
            s = 100, c = 'green', label = 'Iris-virginica')


# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1],   #Centrids for clusters
            s = 100,marker='*', c = 'Blue', label = 'Centroids')
plt.title('Clusters for dataset sepal length and sepal width')
plt.legend()


```

```python
plt.scatter(x[y_kmeans == 0, 2], x[y_kmeans == 0, 3],           # for petal length and petal width
            s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(x[y_kmeans == 1, 2], x[y_kmeans == 1, 3], 
            s = 100, c = 'yellow', label = 'Iris-versicolour')
plt.scatter(x[y_kmeans == 2, 2], x[y_kmeans == 2, 3],
            s = 100, c = 'green', label = 'Iris-virginica')
 
plt.scatter(kmeans.cluster_centers_[:, 2], kmeans.cluster_centers_[:,3],       #centroids for clusters
            s = 100,marker='*', c = 'Blue', label = 'Centroids')
plt.title('Clusters for dataset petal length  and petal  width')
plt.legend()

```
